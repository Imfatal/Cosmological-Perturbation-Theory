```python
from IPython.display import display, Math

# ==========================================
# Manifold and spherical chart
# ==========================================

M = Manifold(4, 'M')

X.<t, r, theta, phi> = M.chart(r't r theta phi')
t, r, theta, phi = X[:]

# ==========================================
# Scale factor
# ==========================================

a_fun = function('a')(t)
a = M.scalar_field(a_fun, name='a')

# ==========================================
# Spatial curvature parameter
# ==========================================

k = var('k')

# ==========================================
# Background FRW metric (curved spatial)
# ==========================================

gbar = M.metric('gbar')

# Time component
gbar[0,0] = -1

# Spatial components
gbar[1,1] = a^2 / (1 - k*r^2)
gbar[2,2] = a^2 * r^2
gbar[3,3] = a^2 * r^2 * sin(theta)^2

# ==========================================
# Levi-Civita connection
# ==========================================

nab_bar = gbar.connection()
Gamma_bar = nab_bar

print("\nNonzero background Christoffel symbols (unique in lower indices):\n")

for l in range(4):
    for m in range(4):
        for n in range(m, 4):   # enforce n >= m to avoid duplicates

            comp = Gamma_bar[l, m, n]

            if not comp.expr().is_zero():

                expr = comp.expr().simplify_full()
                expr_latex = latex(expr)

                # Pretty FRW formatting
                expr_latex = expr_latex.replace(r"a\left(t\right)", "a")
                expr_latex = expr_latex.replace(
                    r"\frac{\partial a}{\partial t}",
                    r"\dot{a}"
                )

                display(Math(
                    rf"\bar{{\Gamma}}^{{{l}}}_{{{m}{n}}} = {expr_latex}"
                ))
```

```python
print("\nNonzero Background Riemann tensor components:\n")

# Compute background Riemann tensor
Riem_bar = gbar.riemann()

for l in range(4):
    for m in range(4):
        for n in range(4):
            for p in range(4):

                comp = Riem_bar[l,m,n,p]

                if not comp.expr().is_zero():

                    expr = comp.expr().simplify_full()
                    expr_latex = latex(expr)

                    # FRW cleanup
                    expr_latex = expr_latex.replace(r"a\left(t\right)", "a")
                    expr_latex = expr_latex.replace(
                        r"\frac{\partial a}{\partial t}",
                        r"\dot{a}"
                    )
                    expr_latex = expr_latex.replace(
                        r"\frac{\partial^2 a}{\partial t^2}",
                        r"\ddot{a}"
                    )

                    latex_str = rf"\bar{{R}}^{{{l}}}_{{{m}{n}{p}}} = {expr_latex}"
                    display(Math(latex_str))
```

```python
print("\nNonzero Background Ricci tensor components (unique):\n")

# Compute background Ricci tensor
Ric_bar = gbar.ricci()

import re

for m in range(4):
    for n in range(m, 4):   # enforce symmetry (n >= m)

        comp = Ric_bar[m,n]

        if not comp.expr().is_zero():

            expr = comp.expr().simplify_full()
            expr_latex = latex(expr)

            # ----------------------------------
            # SAFE coordinate replacements
            # ----------------------------------
            expr_latex = re.sub(r'\bth\b', r'\\theta', expr_latex)
            expr_latex = re.sub(r'\bph\b', r'\\phi', expr_latex)

            # ----------------------------------
            # FRW cleanup
            # ----------------------------------
            expr_latex = expr_latex.replace(r"a\left(t\right)", "a")

            expr_latex = expr_latex.replace(
                r"\frac{\partial a}{\partial t}",
                r"\dot{a}"
            )

            expr_latex = expr_latex.replace(
                r"\frac{\partial^2 a}{\partial t^2}",
                r"\ddot{a}"
            )

            display(Math(
                rf"\bar{{R}}_{{{m}{n}}} = {expr_latex}"
            ))
```

```python
print("\nBackground Energy-Momentum Tensor (Perfect Fluid):\n")

# Define energy density and pressure as scalar fields
rho_fun = function('rho')(t)
p_fun   = function('p')(t)

rho = M.scalar_field(rho_fun, name='rho')
p   = M.scalar_field(p_fun, name='p')

# Define 4-velocity (comoving frame)
u = M.vector_field('u')
u[0] = 1
u[1] = 0
u[2] = 0
u[3] = 0

# Lower index version u_mu
u_cov = u.down(gbar)

# Create background energy-momentum tensor
Tbar = M.tensor_field(0,2, name='Tbar')

for mu in range(4):
    for nu in range(4):
        Tbar[mu,nu] = (rho + p)*u_cov[mu]*u_cov[nu] + p*gbar[mu,nu]

# Print nonzero components
for mu in range(4):
    for nu in range(4):

        comp = Tbar[mu,nu]

        if not comp.expr().is_zero():

            expr = comp.expr().simplify_full()
            expr_latex = latex(expr)

            # Replace symbols with barred versions
            expr_latex = expr_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
            expr_latex = expr_latex.replace(r"p\left(t\right)", r"\bar{p}")
            expr_latex = expr_latex.replace(r"a\left(t\right)", "a")


            latex_str = rf"\bar{{T}}_{{{mu}{nu}}} = {expr_latex}"
            display(Math(latex_str))

print("\nBackground Friedmann Equations (Curved FRW):\n")

display(Math(
    r"\bar{\rho} = \frac{3}{8\pi G}\left("
    r"\frac{\dot a^2}{a^2} + \frac{k}{a^2}"
    r"\right)"
))

display(Math(
    r"\bar{p} = -\frac{1}{8\pi G}\left("
    r"\frac{2\ddot a}{a} + \frac{\dot a^2}{a^2} + \frac{k}{a^2}"
    r"\right)"
))

```

```python
print("\nConservation Equations ∇_μ T̄^{μν} = 0:\n")

# Raise first index
Tbar_up = Tbar.up(gbar, 0)

# Compute divergence
DivTbar = Tbar_up.divergence(gbar)

# Correct spherical coordinate labels
coord_labels = ['t', 'r', r'\theta', r'\phi']

for nu in range(4):

    comp = DivTbar[nu]

    if not comp.expr().is_zero():

        expr = comp.expr().simplify_full()

        expr_latex = latex(expr)

        # Replace background fields
        expr_latex = expr_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
        expr_latex = expr_latex.replace(r"p\left(t\right)", r"\bar{p}")
        expr_latex = expr_latex.replace(r"a\left(t\right)", "a")

        # Dot notation cleanup
        expr_latex = expr_latex.replace(
            r"\frac{\partial \bar{\rho}}{\partial t}",
            r"\dot{\bar{\rho}}"
        )
        expr_latex = expr_latex.replace(
            r"\frac{\partial \bar{p}}{\partial t}",
            r"\dot{\bar{p}}"
        )
        expr_latex = expr_latex.replace(
            r"\frac{\partial a}{\partial t}",
            r"\dot{a}"
        )

        display(Math(
            rf"\nabla_\mu \bar{{T}}^{{\mu {coord_labels[nu]}}} = {expr_latex}"
        ))
```

```python
print("\nTrace-Reversed Background Source Tensor \\bar{S}_{μν}:\n")

# ==========================================
# Compute trace  T̄ = ḡ^{μν} T̄_{μν}
# ==========================================

gbar_inv = gbar.inverse()

T_expr = 0
for mu in range(4):
    for nu in range(4):
        T_expr += gbar_inv[mu,nu].expr() * Tbar[mu,nu].expr()

T_expr = T_expr.simplify_full()

print("Trace \\bar{T} =")

trace_latex = latex(T_expr)

# Replace background fields
trace_latex = trace_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
trace_latex = trace_latex.replace(r"p\left(t\right)", r"\bar{p}")
trace_latex = trace_latex.replace(r"a\left(t\right)", "a")

display(Math(rf"\bar{{T}} = {trace_latex}"))


# ==========================================
# Build S̄_{μν} = T̄_{μν} - 1/2 ḡ_{μν} T̄
# ==========================================

Sbar = M.tensor_field(0,2, name='Sbar')

for mu in range(4):
    for nu in range(4):

        expr = (
            Tbar[mu,nu].expr()
            - (1/2) * gbar[mu,nu].expr() * T_expr
        )

        Sbar[mu,nu] = expr.simplify_full()


# ==========================================
# Print nonzero components
# ==========================================

coord_labels = ['t', 'r', r'\theta', r'\phi']

for mu in range(4):
    for nu in range(4):

        expr = Sbar[mu,nu].expr().simplify_full()

        if not expr.is_zero():

            expr_latex = latex(expr)

            expr_latex = expr_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
            expr_latex = expr_latex.replace(r"p\left(t\right)", r"\bar{p}")
            expr_latex = expr_latex.replace(r"a\left(t\right)", "a")

            display(Math(
                rf"\bar{{S}}_{{{coord_labels[mu]}{coord_labels[nu]}}} = {expr_latex}"
            ))
```

```python
print("\nBackground Einstein Equations (Trace-Reversed Form, unique components):\n")
print(r"\bar{R}_{\mu\nu} = 8\pi G \bar{S}_{\mu\nu}\n")

import re

# -------------------------------------------------
# Newton's constant
# -------------------------------------------------
G = var('G')

# -------------------------------------------------
# Compute background Ricci tensor
# -------------------------------------------------
Ric = gbar.ricci()

# -------------------------------------------------
# Loop over components (enforce symmetry)
# -------------------------------------------------
for mu in range(4):
    for nu in range(mu, 4):   # enforce nu >= mu

        Rcomp = Ric[mu,nu]
        Scomp = Sbar[mu,nu]

        # Skip if both sides zero
        if Rcomp.expr().is_zero() and Scomp.expr().is_zero():
            continue

        # Simplify
        Rexpr = Rcomp.expr().simplify_full()
        Sexpr = (8*pi*G*Scomp.expr()).simplify_full()

        # Convert to LaTeX
        R_latex = latex(Rexpr)
        S_latex = latex(Sexpr)

        # -------------------------------------------------
        # SAFE spherical coordinate replacement
        # -------------------------------------------------
        R_latex = re.sub(r'\bth\b', r'\\theta', R_latex)
        R_latex = re.sub(r'\bph\b', r'\\phi', R_latex)

        S_latex = re.sub(r'\bth\b', r'\\theta', S_latex)
        S_latex = re.sub(r'\bph\b', r'\\phi', S_latex)

        # -------------------------------------------------
        # Replace background quantities
        # -------------------------------------------------
        R_latex = R_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
        R_latex = R_latex.replace(r"p\left(t\right)", r"\bar{p}")
        R_latex = R_latex.replace(r"a\left(t\right)", "a")

        S_latex = S_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
        S_latex = S_latex.replace(r"p\left(t\right)", r"\bar{p}")
        S_latex = S_latex.replace(r"a\left(t\right)", "a")

        # -------------------------------------------------
        # Dot notation cleanup
        # -------------------------------------------------
        R_latex = R_latex.replace(
            r"\frac{\partial a}{\partial t}",
            r"\dot{a}"
        )

        R_latex = R_latex.replace(
            r"\frac{\partial^2 a}{\partial t^2}",
            r"\ddot{a}"
        )

        S_latex = S_latex.replace(
            r"\frac{\partial a}{\partial t}",
            r"\dot{a}"
        )

        display(Math(
            rf"\bar{{R}}_{{{mu}{nu}}} = {R_latex} \quad = \quad {S_latex}"
        ))
```

```python
# ==========================================
# SVT Decomposition (Curved FRW, Spherical)
# ==========================================

print("\nConstructing SVT perturbations (curved spatial FRW)...\n")

# -------------------------------------------------
# Define perturbation tensor
# -------------------------------------------------
h = M.tensor_field(0, 2, name='h')

coords_spatial = [r, theta, phi]

# -------------------------------------------------
# Scalar perturbations
# -------------------------------------------------
A = function('A')(t, r, theta, phi)
B = function('B')(t, r, theta, phi)
E = function('E')(t, r, theta, phi)
F = function('F')(t, r, theta, phi)

# -------------------------------------------------
# Vector perturbations
# -------------------------------------------------
C1 = function('C1')(t, r, theta, phi)
C2 = function('C2')(t, r, theta, phi)
C3 = function('C3')(t, r, theta, phi)

G1 = function('G1')(t, r, theta, phi)
G2 = function('G2')(t, r, theta, phi)
G3 = function('G3')(t, r, theta, phi)

C = [C1, C2, C3]
G = [G1, G2, G3]

# -------------------------------------------------
# Tensor perturbations (TT)
# -------------------------------------------------
D11 = function('D11')(t, r, theta, phi)
D12 = function('D12')(t, r, theta, phi)
D13 = function('D13')(t, r, theta, phi)
D22 = function('D22')(t, r, theta, phi)
D23 = function('D23')(t, r, theta, phi)
D33 = function('D33')(t, r, theta, phi)

D = [
    [D11, D12, D13],
    [D12, D22, D23],
    [D13, D23, D33]
]

# -------------------------------------------------
# Mode switches
# -------------------------------------------------
include_S = False
include_V = False
include_T = True

# ==========================================
# h_00
# ==========================================

if include_S:
    h[0,0] = -E
else:
    h[0,0] = 0

# ==========================================
# h_0i
# ==========================================

for i in range(3):

    comp = 0

    if include_S:
        comp += a * diff(F, coords_spatial[i])

    if include_V:
        comp += a * G[i]

    h[i+1,0] = comp
    h[0,i+1] = comp

# ==========================================
# h_ij
# ==========================================

for i in range(3):
    for j in range(3):

        total = 0

        # ----------------------------
        # Scalar sector
        # ----------------------------
        if include_S:

            # Spatial background metric γ̄_ij = ḡ_ij / a^2
            gamma_ij = gbar[i+1, j+1] / a^2

            # A γ̄_ij term
            total += A * gamma_ij

            # ∂_i ∂_j B (minimal curved version)
            total += diff(diff(B, coords_spatial[i]),
                          coords_spatial[j])

        # ----------------------------
        # Vector sector
        # ----------------------------
        if include_V:
            total += (
                diff(C[j], coords_spatial[i]) +
                diff(C[i], coords_spatial[j])
            )

        # ----------------------------
        # Tensor sector
        # ----------------------------
        if include_T:
            total += D[i][j]

        # Full h_ij
        h[i+1, j+1] = a^2 * total

# ==========================================
# Display nonzero components
# ==========================================

print("\nSVT Decomposed h_{μν} (unique components):\n")

for mu in range(4):
    for nu in range(mu, 4):   # enforce symmetry (nu >= mu)

        comp = h[mu,nu]

        if not comp.expr().is_zero():

            expr = comp.expr().simplify_full()
            display(Math(
                rf"h_{{{mu}{nu}}} = {latex(expr)}"
            ))

```

```python
# ============================================================
# Linearized δΓ in Curved Spherical FRW
# Using background covariant derivative ∇̄
# ============================================================

print("\nComputing δΓ in curved spherical FRW...\n")

# Background inverse metric
gbar_inv = gbar.inverse()

# Background connection
Gamma_bar = nab_bar

# Coordinates
coords = [t, r, theta, phi]

# ------------------------------------------------------------
# Helper: ∇̄_μ h_{αβ}
# ------------------------------------------------------------

def covariant_derivative_h(alpha, beta, mu):
    """
    Computes background covariant derivative:
    ∇̄_μ h_{αβ}
    """

    # Partial derivative
    partial = diff(h[alpha, beta].expr(), coords[mu])

    correction = 0

    for sigma in range(4):

        Gamma1 = Gamma_bar[sigma, mu, alpha].expr()
        Gamma2 = Gamma_bar[sigma, mu, beta].expr()

        # - Γ̄^σ_{μα} h_{σβ}
        correction += Gamma1 * h[sigma, beta].expr()

        # - Γ̄^σ_{μβ} h_{ασ}
        correction += Gamma2 * h[alpha, sigma].expr()

    return partial - correction


# ------------------------------------------------------------
# Build δΓ^μ_{νλ}
# ------------------------------------------------------------

delta_Gamma = M.tensor_field(1,2, name='delta_Gamma')

for mu in range(4):
    for nu in range(4):
        for lam in range(4):

            expr = 0

            for rho in range(4):

                term  = covariant_derivative_h(rho, lam, nu)
                term += covariant_derivative_h(rho, nu, lam)
                term -= covariant_derivative_h(nu, lam, rho)

                expr += gbar_inv[mu, rho].expr() * term

            delta_Gamma[mu,nu,lam] = (1/2) * expr.simplify_full()


# ------------------------------------------------------------
# Display nonzero components (unique in lower indices)
# ------------------------------------------------------------

print("\nNonzero δΓ components (unique in ν,λ):\n")

for mu in range(4):
    for nu in range(4):
        for lam in range(nu, 4):   # enforce λ ≥ ν

            comp = delta_Gamma[mu,nu,lam].expr().simplify_full()

            if not comp.is_zero():

                display(Math(
                    rf"\delta \Gamma^{{{mu}}}_{{{nu}{lam}}} = "
                    + latex(comp)
                ))

```

```python
# ==========================================
# Explicit Linearized δR_{μν}
# (Spherical coordinates, curved FRW)
# ==========================================

def compute_delta_ricci_explicit(delta_Gamma):

    delta_R = M.tensor_field(0,2, name='delta_R')

    # Use already-defined background connection
    Gamma_bar = nab_bar

    coords_all = [t, r, theta, phi]

    for mu in range(4):
        for kappa in range(4):

            expr = 0

            for lam in range(4):

                # ∂_λ δΓ^λ_{μκ}
                expr += diff(
                    delta_Gamma[lam, mu, kappa].expr(),
                    coords_all[lam]
                )

                # − ∂_κ δΓ^λ_{μλ}
                expr -= diff(
                    delta_Gamma[lam, mu, lam].expr(),
                    coords_all[kappa]
                )

                for eta in range(4):

                    Gamma_l_k_e = Gamma_bar[lam, kappa, eta].expr()
                    Gamma_e_m_l = Gamma_bar[eta, mu, lam].expr()
                    Gamma_l_l_e = Gamma_bar[lam, lam, eta].expr()
                    Gamma_e_m_k = Gamma_bar[eta, mu, kappa].expr()

                    # + δΓ^η_{μλ} Γ̄^λ_{κη}
                    expr += (
                        delta_Gamma[eta, mu, lam].expr()
                        * Gamma_l_k_e
                    )

                    # + δΓ^λ_{κη} Γ̄^η_{μλ}
                    expr += (
                        delta_Gamma[lam, kappa, eta].expr()
                        * Gamma_e_m_l
                    )

                    # − δΓ^η_{μκ} Γ̄^λ_{λη}
                    expr -= (
                        delta_Gamma[eta, mu, kappa].expr()
                        * Gamma_l_l_e
                    )

                    # − δΓ^λ_{λη} Γ̄^η_{μκ}
                    expr -= (
                        delta_Gamma[lam, lam, eta].expr()
                        * Gamma_e_m_k
                    )

            delta_R[mu, kappa] = expr.simplify_full()

    return delta_R

# ==========================================
# Compute δR sector-wise
# ==========================================

delta_R = compute_delta_ricci_explicit(delta_Gamma)


# ==========================================
# Print S / V / T contributions (unique)
# ==========================================

print("\nδR decomposed component-wise (S / V / T, unique):\n")

for mu in range(4):
    for nu in range(mu, 4):   # enforce symmetry

        comp = delta_R[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\delta R_{{{mu}{nu}}} = {latex(comp)}"
            ))

```

```python
# ==========================================
# Matter perturbation fields (Curved FRW)
# ==========================================

rho_bar = function('rho_bar',
                   latex_name=r'\bar{\rho}')(t)

p_bar   = function('p_bar',
                   latex_name=r'\bar{p}')(t)

delta_rho = function('delta_rho',
                     latex_name=r'\delta\rho')(t,r,theta,phi)

delta_p = function('delta_p',
                   latex_name=r'\delta p')(t,r,theta,phi)

u_scalar = function('u_scalar',
                    latex_name=r'u^{S}')(t,r,theta,phi)

pi_S = function('pi_S',
                latex_name=r'\pi^{S}')(t,r,theta,phi)

uV = [
    function('uV1', latex_name=r'u^{V}_r')(t,r,theta,phi),
    function('uV2', latex_name=r'u^{V}_\theta')(t,r,theta,phi),
    function('uV3', latex_name=r'u^{V}_\phi')(t,r,theta,phi)
]

piV = [
    function('piV1', latex_name=r'\pi^{V}_r')(t,r,theta,phi),
    function('piV2', latex_name=r'\pi^{V}_\theta')(t,r,theta,phi),
    function('piV3', latex_name=r'\pi^{V}_\phi')(t,r,theta,phi)
]

# Tensor anisotropic stress
piT = [
    [function('piT11', latex_name=r'\pi^{T}_{rr}')(t,r,theta,phi),
     function('piT12', latex_name=r'\pi^{T}_{r\theta}')(t,r,theta,phi),
     function('piT13', latex_name=r'\pi^{T}_{r\phi}')(t,r,theta,phi)],

    [function('piT12', latex_name=r'\pi^{T}_{r\theta}')(t,r,theta,phi),
     function('piT22', latex_name=r'\pi^{T}_{\theta\theta}')(t,r,theta,phi),
     function('piT23', latex_name=r'\pi^{T}_{\theta\phi}')(t,r,theta,phi)],

    [function('piT13', latex_name=r'\pi^{T}_{r\phi}')(t,r,theta,phi),
     function('piT23', latex_name=r'\pi^{T}_{\theta\phi}')(t,r,theta,phi),
     function('piT33', latex_name=r'\pi^{T}_{\phi\phi}')(t,r,theta,phi)]
]

# ==========================================
# Initialize δT tensor
# ==========================================

deltaT = M.tensor_field(0,2, name='deltaT')

for mu in range(4):
    for nu in range(4):
        deltaT[mu,nu] = 0


coords_spatial = [r, theta, phi]

# Spatial metric γ̄_ij = ḡ_ij / a^2
def gamma_bar(i,j):
    return gbar[i+1,j+1] / a^2


# ==========================================
# Build δT
# ==========================================

# ---- SCALAR CONTRIBUTIONS ----
if include_S:

    # δT_00
    deltaT[0,0] = delta_rho

    # δT_0i
    for i in range(3):

        comp = (rho_bar + p_bar) * diff(u_scalar,
                                        coords_spatial[i])

        deltaT[0,i+1] = comp
        deltaT[i+1,0] = comp

    # δT_ij
    for i in range(3):
        for j in range(3):

            scalar_pressure = delta_p * gamma_bar(i,j)

            anisotropic_scalar = diff(
                diff(pi_S, coords_spatial[i]),
                coords_spatial[j]
            )

            deltaT[i+1,j+1] = a^2 * (
                scalar_pressure +
                anisotropic_scalar
            )


# ---- VECTOR CONTRIBUTIONS ----
if include_V:

    for i in range(3):

        comp = (rho_bar + p_bar) * uV[i]

        deltaT[0,i+1] += comp
        deltaT[i+1,0] += comp

    for i in range(3):
        for j in range(3):

            anisotropic_vector = (
                diff(piV[j], coords_spatial[i]) +
                diff(piV[i], coords_spatial[j])
            )

            deltaT[i+1,j+1] += a^2 * anisotropic_vector


# ---- TENSOR CONTRIBUTIONS ----
if include_T:

    for i in range(3):
        for j in range(3):

            deltaT[i+1,j+1] += a^2 * piT[i][j]


# ==========================================
# Metric correction term
# δT_{μν} → δT_{μν} + p̄ h_{μν}
# ==========================================

for mu in range(4):
    for nu in range(4):

        if not h[mu,nu].expr().is_zero():

            deltaT[mu,nu] += p_bar * h[mu,nu]


# ==========================================
# Print δT
# ==========================================

print("\nδT_{μν} components (Curved FRW):\n")

for mu in range(4):
    for nu in range(4):

        comp = deltaT[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\delta T_{{{mu}{nu}}} = "
                + latex(comp)
            ))
```

```python
print("\n--------------------------------------------")
print("Mixed Components  δT^μ_ν  (S / V / T)")
print("--------------------------------------------\n")

# Background inverse metric
gbar_inv = gbar.inverse()

# Raise first index properly
deltaT_mixed = gbar.inverse().contract(1, deltaT, 0)


print("\n--- δT^μ_ν decomposed ---\n")

for mu in range(4):
    for nu in range(4):

        comp = deltaT_mixed[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\delta T^{{{mu}}}_{{{nu}}} = "
                + latex(comp)
            ))


print("\n--------------------------------------------")
print("Trace  δT^λ_λ  (S / V / T)")
print("--------------------------------------------\n")

trace = 0
for lam in range(4):
    trace += deltaT_mixed[lam,lam].expr()

trace = trace.simplify_full()

display(Math(
    r"\delta T^{\lambda}_{\ \lambda} = " + latex(trace)
))

```

```python
print("\n--------------------------------------------")
print("Perturbed Source Tensor  δS_{μν}")
print("--------------------------------------------\n")

# -------------------------------------------------
# Background trace  T̄ = ḡ^{μν} T̄_{μν}
# For perfect fluid:  T̄ = -ρ̄ + 3p̄
# -------------------------------------------------

Tbar_expr = (-rho_bar + 3*p_bar).simplify_full()

# -------------------------------------------------
# Raise first index of δT
# -------------------------------------------------

gbar_inv = gbar.inverse()
deltaT_mixed = gbar_inv.contract(1, deltaT, 0)

# -------------------------------------------------
# Compute trace  δT^λ_λ  explicitly
# -------------------------------------------------

trace_expr = 0
for lam in range(4):
    trace_expr += deltaT_mixed[lam,lam].expr()

trace_expr = trace_expr.simplify_full()

# -------------------------------------------------
# Build δS_{μν}
# δS_{μν} = δT_{μν}
#           − 1/2 ḡ_{μν} δT
#           − 1/2 h_{μν} T̄
# -------------------------------------------------

deltaS = M.tensor_field(0,2, name='deltaS')

for mu in range(4):
    for nu in range(4):

        expr = (
            deltaT[mu,nu].expr()
            - (1/2)*gbar[mu,nu].expr()*trace_expr
            - (1/2)*h[mu,nu].expr()*Tbar_expr
        )

        deltaS[mu,nu] = expr.simplify_full()

# -------------------------------------------------
# Print nonzero components
# -------------------------------------------------

print("\n--- δS_{μν} components ---\n")

for mu in range(4):
    for nu in range(4):

        comp = deltaS[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\delta S_{{{mu}{nu}}} = "
                + latex(comp)
            ))
```

```python
print("\n============================================")
print("Linearized Einstein Equations")
print("============================================\n")

# -------------------------------------------------
# Newton constant
# -------------------------------------------------

Gsym = var('G')
eight_pi_G = 8*pi*Gsym

# -------------------------------------------------
# Initialize Einstein tensor
# -------------------------------------------------

Ein = M.tensor_field(0,2, name='Ein')

# -------------------------------------------------
# Construct equations
# 𝓔_{μν} = δR_{μν} − 8πG δS_{μν}
# -------------------------------------------------

for mu in range(4):
    for nu in range(mu, 4):   # <-- only upper triangle

        expr = (
            delta_R[mu,nu].expr()
            - eight_pi_G * deltaS[mu,nu].expr()
        )

        Ein[mu,nu] = expr.simplify_full()

print("\nEinstein equations constructed.\n")

print("\n--- δR_{μν} − 8πG δS_{μν} ---\n")

for mu in range(4):
    for nu in range(mu, 4):   # only upper triangle (avoids duplicates)

        comp = Ein[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\mathcal{{E}}_{{{mu}{nu}}} = "
                + latex(comp)
            ))

```
