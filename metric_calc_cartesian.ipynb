```python
from IPython.display import display, Math

# Manifold and spherical chart
M = Manifold(4, 'M')

X.<t, x, y, z> = M.chart(r't x y z')
t, x, y, z = X[:]

# Scale factor
a_fun = function('a')(t)
a = M.scalar_field(a_fun, name='a')

# Spatial curvature parameter
k = var('k')

# Background metric
gbar = M.metric('gbar')

# Time component
gbar[0,0] = -1

# Spatial components (general k)
# gbar[1,1] = a^2 / (1 - k*r^2)
# gbar[2,2] = a^2 * r^2
# gbar[3,3] = a^2 * r^2 * sin(th)^2

gbar[1,1] = a^2 
gbar[2,2] = a^2
gbar[3,3] = a^2

# Levi-Civita connection
nab_bar = gbar.connection()
Gamma_bar = nab_bar

print("\nNonzero background Christoffel symbols (unique in lower indices):\n")

for l in range(4):
    for m in range(4):
        for n in range(m, 4):   # <-- enforce n >= m to avoid duplicates

            comp = Gamma_bar[l, m, n]

            if not comp.expr().is_zero():

                expr = comp.expr().simplify_full()
                expr_latex = latex(expr)

                expr_latex = expr_latex.replace(r"a\left(t\right)", "a")
                expr_latex = expr_latex.replace(
                    r"\frac{\partial a}{\partial t}",
                    r"\dot{a}"
                )

                display(Math(
                    rf"\bar{{\Gamma}}^{{{l}}}_{{{m}{n}}} = {expr_latex}"
                ))

```

```python
print("\nNonzero Background Riemann tensor components:\n")

# Compute background Riemann tensor
Riem_bar = gbar.riemann()

for l in range(4):
    for m in range(4):
        for n in range(4):
            for p in range(4):

                comp = Riem_bar[l,m,n,p]

                if not comp.expr().is_zero():

                    expr = comp.expr().simplify_full()
                    expr_latex = latex(expr)

                    # FRW cleanup
                    expr_latex = expr_latex.replace(r"a\left(t\right)", "a")
                    expr_latex = expr_latex.replace(
                        r"\frac{\partial a}{\partial t}",
                        r"\dot{a}"
                    )
                    expr_latex = expr_latex.replace(
                        r"\frac{\partial^2 a}{\partial t^2}",
                        r"\ddot{a}"
                    )

                    latex_str = rf"\bar{{R}}^{{{l}}}_{{{m}{n}{p}}} = {expr_latex}"
                    display(Math(latex_str))
```

```python
print("\nNonzero Background Ricci tensor components (unique):\n")

Ric_bar = gbar.ricci()

for m in range(4):
    for n in range(m, 4):   # enforce symmetry

        comp = Ric_bar[m,n]

        if not comp.expr().is_zero():

            expr = comp.expr().simplify_full()
            expr_latex = latex(expr)

            # FRW cleanup
            expr_latex = expr_latex.replace(r"a\left(t\right)", "a")
            expr_latex = expr_latex.replace(
                r"\frac{\partial a}{\partial t}",
                r"\dot{a}"
            )
            expr_latex = expr_latex.replace(
                r"\frac{\partial^2 a}{\partial t^2}",
                r"\ddot{a}"
            )

            display(Math(
                rf"\bar{{R}}_{{{m}{n}}} = {expr_latex}"
            ))
```

```python
print("\nBackground Ricci scalar:\n")

# Compute Ricci scalar
Rscalar = gbar.ricci_scalar()

expr = Rscalar.expr().simplify_full()
expr_latex = latex(expr)

# Optional FRW cleanup
expr_latex = expr_latex.replace(r"a\left(t\right)", "a")
expr_latex = expr_latex.replace(
    r"\frac{\partial a}{\partial t}",
    r"\dot{a}"
)
expr_latex = expr_latex.replace(
    r"\frac{\partial^2 a}{\partial t^2}",
    r"\ddot{a}"
)

display(Math(rf"\bar{{R}} = {expr_latex}"))

```

```python
print("\nBackground Energy-Momentum Tensor (Perfect Fluid):\n")

# Define energy density and pressure as scalar fields
rho_fun = function('rho')(t)
p_fun   = function('p')(t)

rho = M.scalar_field(rho_fun, name='rho')
p   = M.scalar_field(p_fun, name='p')

# Define 4-velocity (comoving frame)
u = M.vector_field('u')
u[0] = 1
u[1] = 0
u[2] = 0
u[3] = 0

# Lower index version u_mu
u_cov = u.down(gbar)

# Create background energy-momentum tensor
Tbar = M.tensor_field(0,2, name='Tbar')

for mu in range(4):
    for nu in range(4):
        Tbar[mu,nu] = (rho + p)*u_cov[mu]*u_cov[nu] + p*gbar[mu,nu]

# Print nonzero components
for mu in range(4):
    for nu in range(4):

        comp = Tbar[mu,nu]

        if not comp.expr().is_zero():

            expr = comp.expr().simplify_full()
            expr_latex = latex(expr)

            # Replace symbols with barred versions
            expr_latex = expr_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
            expr_latex = expr_latex.replace(r"p\left(t\right)", r"\bar{p}")
            expr_latex = expr_latex.replace(r"a\left(t\right)", "a")


            latex_str = rf"\bar{{T}}_{{{mu}{nu}}} = {expr_latex}"
            display(Math(latex_str))


print("\nBackground Friedmann Equations:\n")

display(Math(
    r"\bar{\rho} = \frac{3}{8\pi G}\left(\frac{\dot a^2}{a^2}\right)"
))

display(Math(
    r"\bar{p} = -\frac{1}{8\pi G}\left("
    r"\frac{2\ddot a}{a} + \frac{\dot a^2}{a^2}"
    r"\right)"
))
```

```python
print("\nConservation Equations ∇_μ T̄^{μν} = 0:\n")

# Raise first index
Tbar_up = Tbar.up(gbar, 0)

# Compute divergence
DivTbar = Tbar_up.divergence(gbar)

# Coordinate labels
coord_labels = ['t', 'x', 'y', 'z']

for nu in range(4):

    comp = DivTbar[nu]

    if not comp.expr().is_zero():

        expr = comp.expr().simplify_full()
        expr_latex = latex(expr)

        # Replace time-dependent fields with barred versions
        expr_latex = expr_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
        expr_latex = expr_latex.replace(r"p\left(t\right)", r"\bar{p}")
        expr_latex = expr_latex.replace(r"a\left(t\right)", "a")

        # Dot notation
        expr_latex = expr_latex.replace(
            r"\frac{\partial \bar{\rho}}{\partial t}",
            r"\dot{\bar{\rho}}"
        )
        expr_latex = expr_latex.replace(
            r"\frac{\partial \bar{p}}{\partial t}",
            r"\dot{\bar{p}}"
        )
        expr_latex = expr_latex.replace(
            r"\frac{\partial a}{\partial t}",
            r"\dot{a}"
        )

        # Use coordinate names instead of numbers
        nu_label = coord_labels[nu]

        latex_str = rf"\nabla_\mu \bar{{T}}^{{\mu {nu_label}}} = {expr_latex}"
        display(Math(latex_str))
```

```python
print("\nTrace-Reversed Background Source Tensor \\bar{S}_{μν}:\n")

# --- Compute trace T̄ = ḡ^{μν} T̄_{μν} ---

gbar_inv = gbar.inverse()

T_expr = 0
for mu in range(4):
    for nu in range(4):
        T_expr += gbar_inv[mu,nu].expr() * Tbar[mu,nu].expr()

T_expr = T_expr.simplify_full()


print("Trace \\bar{T} =")

trace_latex = latex(T_expr)

trace_latex = trace_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
trace_latex = trace_latex.replace(r"p\left(t\right)", r"\bar{p}")
trace_latex = trace_latex.replace(r"a\left(t\right)", r"\bar{a}")

display(Math(rf"\bar{{T}} = {trace_latex}"))



# --- Build S̄_{μν} ---

Sbar = M.tensor_field(0,2, name='Sbar')

for mu in range(4):
    for nu in range(4):

        expr = Tbar[mu,nu].expr() \
               - (1/2) * gbar[mu,nu].expr() * T_expr

        Sbar[mu,nu] = expr


# --- Print nonzero components ---

for mu in range(4):
    for nu in range(4):

        expr = Sbar[mu,nu].expr().simplify_full()

        if not expr.is_zero():

            expr_latex = latex(expr)

            expr_latex = expr_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
            expr_latex = expr_latex.replace(r"p\left(t\right)", r"\bar{p}")
            expr_latex = expr_latex.replace(r"a\left(t\right)", r"\bar{a}")

            display(Math(rf"\bar{{S}}_{{{mu}{nu}}} = {expr_latex}"))
```

```python
print("\nBackground Einstein Equations (Trace-Reversed Form, unique components):\n")
print(r"\bar{R}_{\mu\nu} = 8\pi G \bar{S}_{\mu\nu}\n")

G = var('G', latex_name=r'\bar{G}')

Ric = gbar.ricci()

for mu in range(4):
    for nu in range(mu, 4):   # enforce symmetry (nu >= mu)

        Rcomp = Ric[mu,nu]
        Scomp = Sbar[mu,nu]

        # Skip if both sides are zero
        if Rcomp.expr().is_zero() and Scomp.expr().is_zero():
            continue

        # Simplify expressions
        Rexpr = Rcomp.expr().simplify_full()
        Sexpr = (8*pi*G*Scomp.expr()).simplify_full()

        # Convert to LaTeX
        R_latex = latex(Rexpr)
        S_latex = latex(Sexpr)

        # Clean formatting
        R_latex = R_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
        R_latex = R_latex.replace(r"p\left(t\right)", r"\bar{p}")
        R_latex = R_latex.replace(r"a\left(t\right)", r"a")

        S_latex = S_latex.replace(r"\rho\left(t\right)", r"\bar{\rho}")
        S_latex = S_latex.replace(r"p\left(t\right)", r"\bar{p}")
        S_latex = S_latex.replace(r"a\left(t\right)", r"a")

        display(Math(
            rf"\bar{{R}}_{{{mu}{nu}}} = {R_latex} \quad = \quad {S_latex}"
        ))
```

```python
# ==========================================
# SVT Decomposition (Mukhanov-style vars)
# ==========================================

h = M.tensor_field(0, 2, name='h')

# ----- Scalar fields -----
A  = function('A')(t, x, y, z)
B  = function('B')(t, x, y, z)
E  = function('E')(t, x, y, z)
F  = function('F')(t, x, y, z)

# ----- Vector fields -----
C1 = function('C1')(t, x, y, z)
C2 = function('C2')(t, x, y, z)
C3 = function('C3')(t, x, y, z)

G1 = function('G1')(t, x, y, z)
G2 = function('G2')(t, x, y, z)
G3 = function('G3')(t, x, y, z)

C = [C1, C2, C3]
G = [G1, G2, G3]

# ----- Tensor modes (TT) -----
D11 = function('D11')(t, x, y, z)
D12 = function('D12')(t, x, y, z)
D13 = function('D13')(t, x, y, z)
D22 = function('D22')(t, x, y, z)
D23 = function('D23')(t, x, y, z)
D33 = function('D33')(t, x, y, z)

D = [
    [D11, D12, D13],
    [D12, D22, D23],
    [D13, D23, D33]
]

coords = [x, y, z]
delta = lambda i,j: 1 if i==j else 0

# ===========================
# Assign components
# ===========================

# ==========================================
# Mode switches
# ==========================================

include_S = False
include_V = True
include_T = False

h = M.tensor_field(0, 2, name='h')

coords = [x, y, z]
delta = lambda i,j: 1 if i==j else 0

# ===========================
# h_00
# ===========================

if include_S:
    h[0,0] = -E
else:
    h[0,0] = 0

# ===========================
# h_i0
# ===========================

for i in range(3):

    comp = 0

    if include_S:
        comp += a * diff(F, coords[i])

    if include_V:
        comp += a * G[i]

    h[i+1,0] = comp
    h[0,i+1] = comp

# ===========================
# h_ij
# ===========================

for i in range(3):
    for j in range(3):

        total = 0

        if include_S:
            scalar_part = A * delta(i,j)
            B_part = diff(diff(B, coords[i]), coords[j])
            total += scalar_part + B_part

        if include_V:
            C_part = diff(C[j], coords[i]) + diff(C[i], coords[j])
            total += C_part

        if include_T:
            total += D[i][j]

        h[i+1, j+1] = a^2 * total



print("\nSVT Decomposed h_{mu nu}:\n")

for mu in range(4):
    for nu in range(4):
        comp = h[mu,nu]
        if not comp.expr().is_zero():
            display(Math(rf"h_{{{mu}{nu}}} = {latex(comp.expr())}"))
```

```python
# ==========================================
# SVT Constraint Equations
# ==========================================

# --- Divergences ---

div_C = diff(C1, x) + diff(C2, y) + diff(C3, z)
div_G = diff(G1, x) + diff(G2, y) + diff(G3, z)

div_D = [
    diff(D11, x) + diff(D12, y) + diff(D13, z),
    diff(D12, x) + diff(D22, y) + diff(D23, z),
    diff(D13, x) + diff(D23, y) + diff(D33, z)
]

trace_D = D11 + D22 + D33

print("SVT Constraints:\n")

# --- Display nicely with = 0 ---

display(Math(
    r"\partial_i C_i = "
    + latex(div_C)
    + r" = 0"
))

display(Math(
    r"\partial_i G_i = "
    + latex(div_G)
    + r" = 0"
))

display(Math(
    r"\partial_i D_{ij} = "
    + latex(div_D)
    + r" = 0"
))

display(Math(
    r"D_{ii} = "
    + latex(trace_D)
    + r" = 0"
))

# ==========================================
# Substitution Rules (for later simplification)
# ==========================================

constraint_rules = {
    div_C: 0,
    div_G: 0,
    trace_D: 0,
    div_D[0]: 0,
    div_D[1]: 0,
    div_D[2]: 0
}
```

```python
# ==========================================
# Correct Linearized δΓ using covariant derivatives
# ==========================================

def compute_delta_gamma_explicit(h):

    delta_Gamma = M.tensor_field(1,2)
    gbar_inv = gbar.inverse()

    a_expr = a.expr()
    a_dot = diff(a_expr, t)

    coords = [t, x, y, z]

    # --------------------------------------
    # Background Christoffel Γ̄^α_{βγ}
    # --------------------------------------
    def Gamma_bar(alpha, beta, gamma):

        # Γ̄^0_{ij}
        if alpha == 0 and beta > 0 and gamma > 0:
            if beta == gamma:
                return a_expr * a_dot

        # Γ̄^i_{j0} and Γ̄^i_{0j}
        if alpha > 0:
            if (gamma == 0 and beta > 0 and alpha == beta):
                return a_dot / a_expr
            if (beta == 0 and gamma > 0 and alpha == gamma):
                return a_dot / a_expr

        return 0

    # --------------------------------------
    # Covariant derivative of h
    # ∇̄_ν h_{ρλ}
    # --------------------------------------
    def covD(nu, rho, lam):

        term = diff(h[rho, lam].expr(), coords[nu])

        for sigma in range(4):

            term -= Gamma_bar(sigma, nu, rho) * h[sigma, lam].expr()
            term -= Gamma_bar(sigma, nu, lam) * h[rho, sigma].expr()

        return term

    # --------------------------------------
    # Build δΓ
    # --------------------------------------
    for mu in range(4):
        for nu in range(4):
            for lam in range(4):

                expr = 0

                for rho in range(4):

                    term = (
                        covD(nu, rho, lam)
                        + covD(lam, rho, nu)
                        - covD(rho, nu, lam)
                    )

                    expr += gbar_inv[mu, rho].expr() * term

                delta_Gamma[mu, nu, lam] = (1/2) * expr.simplify_full()

    return delta_Gamma


# ==========================================
# Compute δΓ for active sector (single h)
# ==========================================

delta_Gamma = compute_delta_gamma_explicit(h)

print("\nδΓ components (unique in lower indices):\n")

for mu in range(4):
    for nu in range(4):
        for lam in range(nu, 4):   # enforce lam >= nu

            comp = delta_Gamma[mu,nu,lam].expr()

            if not comp.is_zero():

                display(Math(
                    rf"\delta \Gamma^{{{mu}}}_{{{nu}{lam}}}"
                ))

                display(Math(
                    r"\quad = " + latex(comp)
                ))

                display(Math(r"\qquad"))


```

```python
def compute_delta_ricci_clean(delta_Gamma):

    delta_R = M.tensor_field(0,2)
    coords = [t, x, y, z]

    a_expr = a.expr()
    a_dot = diff(a_expr, t)

    def Gamma_bar(alpha, beta, gamma):

        if alpha == 0 and beta > 0 and gamma > 0:
            if beta == gamma:
                return a_expr * a_dot

        if alpha > 0:
            if gamma == 0 and beta > 0 and alpha == beta:
                return a_dot / a_expr
            if beta == 0 and gamma > 0 and alpha == gamma:
                return a_dot / a_expr

        return 0

    for mu in range(4):
        for nu in range(4):

            expr = 0

            for lam in range(4):

                # ∂_λ δΓ^λ_{μν}
                expr += diff(delta_Gamma[lam, mu, nu].expr(), coords[lam])

                # − ∂_ν δΓ^λ_{μλ}
                expr -= diff(delta_Gamma[lam, mu, lam].expr(), coords[nu])

                for rho in range(4):

                    # + Γ̄^λ_{λρ} δΓ^ρ_{μν}
                    expr += Gamma_bar(lam, lam, rho) * delta_Gamma[rho, mu, nu].expr()

                    # + Γ̄^ρ_{μν} δΓ^λ_{λρ}
                    expr += Gamma_bar(rho, mu, nu) * delta_Gamma[lam, lam, rho].expr()

                    # − Γ̄^ρ_{λν} δΓ^λ_{μρ}
                    expr -= Gamma_bar(rho, lam, nu) * delta_Gamma[lam, mu, rho].expr()

                    # − Γ̄^ρ_{μλ} δΓ^λ_{ρν}
                    expr -= Gamma_bar(rho, mu, lam) * delta_Gamma[lam, rho, nu].expr()

            delta_R[mu, nu] = expr.simplify_full()

    return delta_R


# ==========================================
# Display δR (unique components only)
# ==========================================

delta_R = compute_delta_ricci_clean(delta_Gamma)

print("\nδR components (unique):\n")

for mu in range(4):
    for nu in range(mu, 4):   # enforce symmetry

        comp = delta_R[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\delta R_{{{mu}{nu}}} = "
                + latex(comp)
            ))

```

```python
# ==========================================
# Matter perturbation fields
# ==========================================

rho_bar = function('rho_bar', latex_name=r'\bar{\rho}')(t)
p_bar   = function('p_bar',   latex_name=r'\bar{p}')(t)

delta_rho = function('delta_rho',
                     latex_name=r'\delta\rho')(t,x,y,z)

delta_p = function('delta_p',
                   latex_name=r'\delta p')(t,x,y,z)

u_scalar = function('u_scalar',
                    latex_name=r'u^{S}')(t,x,y,z)

pi_S = function('pi_S',
                latex_name=r'\pi^{S}')(t,x,y,z)

uV = [
    function('uV1', latex_name=r'u^{V}_1')(t,x,y,z),
    function('uV2', latex_name=r'u^{V}_2')(t,x,y,z),
    function('uV3', latex_name=r'u^{V}_3')(t,x,y,z)
]


# piV = [
#     function('piV1', latex_name=r'\pi^{V}_1')(t,x,y,z),
#     function('piV2', latex_name=r'\pi^{V}_2')(t,x,y,z),
#     function('piV3', latex_name=r'\pi^{V}_3')(t,x,y,z)
# ]

# For a perfect fluid piV = 0
piV = [
    0, 
    0, 
    0
]

piT = [
    [function('piT11', latex_name=r'\pi^{T}_{11}')(t,x,y,z),
     function('piT12', latex_name=r'\pi^{T}_{12}')(t,x,y,z),
     function('piT13', latex_name=r'\pi^{T}_{13}')(t,x,y,z)],

    [function('piT12', latex_name=r'\pi^{T}_{12}')(t,x,y,z),
     function('piT22', latex_name=r'\pi^{T}_{22}')(t,x,y,z),
     function('piT23', latex_name=r'\pi^{T}_{23}')(t,x,y,z)],

    [function('piT13', latex_name=r'\pi^{T}_{13}')(t,x,y,z),
     function('piT23', latex_name=r'\pi^{T}_{23}')(t,x,y,z),
     function('piT33', latex_name=r'\pi^{T}_{33}')(t,x,y,z)]
]


# ==========================================
# Initialize δT tensor
# ==========================================

deltaT = M.tensor_field(0,2)
# Initialize all components to zero explicitly
for mu in range(4):
    for nu in range(4):
        deltaT[mu,nu] = 0


coords_spatial = [x,y,z]
delta = lambda i,j: 1 if i==j else 0


# ==========================================
# Build δT according to active sector
# ==========================================

# ---- SCALAR CONTRIBUTIONS ----
if include_S:

    # δT_00
    deltaT[0,0] = delta_rho

    # δT_0i
    for i in range(3):
        comp = (rho_bar + p_bar) * diff(u_scalar, coords_spatial[i])
        deltaT[0,i+1] = comp
        deltaT[i+1,0] = comp

    # δT_ij
    for i in range(3):
        for j in range(3):

            scalar_pressure = delta_p * delta(i,j)
            anisotropic_scalar = diff(diff(pi_S, coords_spatial[i]),
                                      coords_spatial[j])

            deltaT[i+1,j+1] = a^2 * (
                scalar_pressure + anisotropic_scalar
            )


# ---- VECTOR CONTRIBUTIONS ----
if include_V:

    for i in range(3):

        comp = (rho_bar + p_bar) * uV[i]
        deltaT[0,i+1] += comp
        deltaT[i+1,0] += comp

    for i in range(3):
        for j in range(3):

            anisotropic_vector = (
                diff(piV[j], coords_spatial[i]) +
                diff(piV[i], coords_spatial[j])
            )

            deltaT[i+1,j+1] += a^2 * anisotropic_vector


# ---- TENSOR CONTRIBUTIONS ----
if include_T:

    for i in range(3):
        for j in range(3):

            deltaT[i+1,j+1] += a^2 * piT[i][j]


# ==========================================
# Metric correction term
# δT_{μν} → δT_{μν} + p̄ h_{μν}
# ==========================================

for mu in range(4):
    for nu in range(4):

        if not h[mu,nu].expr().is_zero():
            deltaT[mu,nu] += p_bar * h[mu,nu]


# ==========================================
# Print δT
# ==========================================

print("\nδT_{μν} components:\n")

for mu in range(4):
    for nu in range(4):

        comp = deltaT[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\delta T_{{{mu}{nu}}} = "
                + latex(comp)
            ))
```

```python
print("\n--------------------------------------------")
print("Mixed Components  δT^μ_ν")
print("--------------------------------------------\n")

# Background inverse metric
gbar_inv = gbar.inverse()

# Raise first index
deltaT_mixed = gbar_inv.contract(1, deltaT, 0)

print("\n--- δT^μ_ν ---\n")

for mu in range(4):
    for nu in range(4):

        comp = deltaT_mixed[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(rf"\delta T^{{{mu}}}_{{{nu}}}"))
            display(Math(r"\quad = " + latex(comp)))
            display(Math(r"\qquad"))

print("\n--------------------------------------------")
print("Trace  δT^λ_λ")
print("--------------------------------------------\n")

trace = deltaT_mixed.trace().expr().simplify_full()

display(Math(r"\delta T^{\lambda}_{\ \lambda}"))

if not trace.is_zero():
    display(Math(r"\quad = " + latex(trace)))
```

```python
print("\n--------------------------------------------")
print("Perturbed Source Tensor  δS_{μν}")
print("--------------------------------------------\n")

# Background trace
Tbar_trace = -rho_bar + 3*p_bar

# Use previously computed mixed tensor
gbar_inv = gbar.inverse()
deltaT_mixed = gbar_inv.contract(1, deltaT, 0)

# Trace δT^λ_λ
trace = deltaT_mixed.trace().expr().simplify_full()

# Initialize δS
deltaS = M.tensor_field(0,2)

for mu in range(4):
    for nu in range(4):

        deltaS[mu,nu] = (
            deltaT[mu,nu]
            - (1/2)*gbar[mu,nu]*trace
            - (1/2)*h[mu,nu]*Tbar_trace
        )


print("\n--- δS_{μν} components ---\n")

for mu in range(4):
    for nu in range(4):

        comp = deltaS[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\delta S_{{{mu}{nu}}} = "
                + latex(comp)
            ))
```

```python
print("\n============================================")
print("Linearized Einstein Equations")
print("============================================\n")

# Newton constant
Gsym = var('G')
eight_pi_G = 8*pi*Gsym

# Initialize Einstein tensor
Ein = M.tensor_field(0,2)

for mu in range(4):
    for nu in range(4):

        Rexpr = delta_R[mu,nu].expr()
        Sexpr = deltaS[mu,nu].expr()

        Ein[mu,nu] = (Rexpr - eight_pi_G * Sexpr).simplify_full()


print("\nEinstein equations constructed.\n")


# ==========================================
# Print components
# ==========================================

print("\n--- δR_{μν} - 8πG δS_{μν} ---\n")

for mu in range(4):
    for nu in range(mu, 4):   # <-- only upper triangle

        comp = Ein[mu,nu].expr().simplify_full()

        if not comp.is_zero():

            display(Math(
                rf"\mathcal{{E}}_{{{mu}{nu}}} = "
                + latex(comp)
            ))
```

```python
print("\n============================================")
print("Perturbed Conservation Equations")
print("============================================\n")

coords = [t, x, y, z]

# Ensure background mixed tensor exists
Tbar_mixed = gbar.inverse().contract(1, Tbar, 0)

# Ensure deltaT_mixed exists
deltaT_mixed = gbar.inverse().contract(1, deltaT, 0)

Cons = [0,0,0,0]

for nu in range(4):

    expr = 0

    # =====================================
    # 1) ∂_μ δT^μ_ν
    # =====================================
    for mu in range(4):
        expr += diff(deltaT_mixed[mu,nu].expr(), coords[mu])

    # =====================================
    # 2) + Γ̄^μ_{μλ} δT^λ_ν
    # =====================================
    for mu in range(4):
        for lam in range(4):
            expr += (
                Gamma_bar[mu,mu,lam].expr()
                * deltaT_mixed[lam,nu].expr()
            )

    # =====================================
    # 3) − Γ̄^λ_{μν} δT^μ_λ
    # =====================================
    for mu in range(4):
        for lam in range(4):
            expr -= (
                Gamma_bar[lam,mu,nu].expr()
                * deltaT_mixed[mu,lam].expr()
            )

    # =====================================
    # 4) + δΓ^μ_{μλ} T̄^λ_ν
    # =====================================
    for mu in range(4):
        for lam in range(4):
            expr += (
                delta_Gamma[mu,mu,lam].expr()
                * Tbar_mixed[lam,nu].expr()
            )

    # =====================================
    # 5) − δΓ^λ_{μν} T̄^μ_λ
    # =====================================
    for mu in range(4):
        for lam in range(4):
            expr -= (
                delta_Gamma[lam,mu,nu].expr()
                * Tbar_mixed[mu,lam].expr()
            )

    Cons[nu] = expr.simplify_full()


# ==========================================
# Printing logic
# ==========================================

if include_S:

    print("\n--- Energy Equation (ν = 0) ---\n")
    if not Cons[0].is_zero():
        display(Math(r"\mathcal{C}_0 = " + latex(Cons[0])))

    print("\n--- Momentum Equation (ν = i) ---\n")
    for i in range(1,4):
        if not Cons[i].is_zero():
            display(Math(rf"\mathcal{{C}}_{i} = " + latex(Cons[i])))

elif include_V:

    print("\n--- Momentum Equation (Vector Sector) ---\n")
    for i in range(1,4):
        if not Cons[i].is_zero():
            display(Math(rf"\mathcal{{C}}_{i} = " + latex(Cons[i])))

elif include_T:

    print("\nTensor sector: no conservation equation.\n")
```
